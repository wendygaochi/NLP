```python
import re
import pandas as pd
```

```python
df_inspect = pd.read_csv("Food_Inspections.csv")
```

```python
df_inspect.head()
```

```python
df_fail = df_inspect[df_inspect['Results'] == 'Fail']
df_fail = df_fail.dropna()
```

```python
df_fail.head()
```

```python
df = df_fail['Violations']
df.head(10)
```

```python
text = '3. POTENTIALLY HAZARDOUS'
re.findall(r'^[0-9]+', text)
```

```python
dic = {}
for d in df.values:
    if not pd.isnull(d): # filter out nan in the series
        items = d.split(' | ') # causes are separated by |
        for text in items: # for each cause
            if re.match(r'^[0-9]+\.', text): # the cause must start with some numbers with a dot
                cause = text.split(' - Comments')[0] # remove the comments in the cause
                if cause in dic: dic[cause] += 1
                else: dic[cause] = 1
```

```python
# sort the causes with respect to the counts
for key, value in sorted(dic.items(), key=lambda k_v: k_v[1], reverse=True)[:10]:
    print ("%s: %s" % (key, value))
```
